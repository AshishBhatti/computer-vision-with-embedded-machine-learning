```python
import numpy as np  # Free Library for matrix manipulations
import matplotlib.pyplot as plt    # To view our matrix as image
import PIL   # Python IMaging Library, Read and write images to files
```

```python
# Load location of image
image_path = "/content/elephant_5x4_gray.bmp"
```

```python
# Open Image (as grayscale) and convert to numpy array
img = PIL.Image.open(image_path).convert('L')
print(img)
```

```python
# Convert to numpy array
img_array = np.asarray(img)
print(img_array.shape)
print("(num rows, num columns)")
print()
print(img_array)
```

```python
# Show normalized array (values between 0 and 1)
print(img_array / 255)
```

```python
# Display image from array
plt.imshow(img_array, cmap='gray', vmin=0, vmax=255)
```

```python
img_vector = img_array.reshape(1,20)
print(img_vector)
```

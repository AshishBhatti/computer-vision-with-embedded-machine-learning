```python
import numpy as np
import matplotlib.pyplot as plt
import PIL
```

```python
image_path = "/content/sea_turtle_5x4_color.bmp"
```

```python
# Open image as color and convert to numpy array
img = PIL.Image.open(image_path).convert('RGB')
print(img)
```

```python
#convert to numpy array
img_array  = np.asarray(img)
print(img_array.shape)
print(img_array)
```

```python
# Display image channels separately

#create 3 copies of array
img_r = img_array.copy()
img_g = img_array.copy()
img_b = img_array.copy()

# Fill non red chanels with 0
img_r[:,:,1].fill(0)
img_r[:,:,2].fill(0)

# Fill non green chanels with 0
img_g[:,:,0].fill(0)
img_g[:,:,2].fill(0)

# Fill non blue chanels with 0
img_b[:,:,0].fill(0)
img_b[:,:,1].fill(0)

#Display each of these channels separately
plt.figure()
plt.imshow(img_r, vmin=0, vmax=255)
plt.figure()
plt.imshow(img_g, vmin=0, vmax=255)
plt.figure()
plt.imshow(img_b, vmin=0, vmax=255)
```

```python
# Show all channels together
plt.imshow(img_array, vmin=0 , vmax=255)
```
